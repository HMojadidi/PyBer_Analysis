```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot for a Line Chart in MATLAB:
plt.plot(x_axis, y_axis)
```

```python
# Create the plot for a Line Chart with Object-Oriented Interface with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Annotating the Chart in MATLAB: Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot (Still Annotating the Chart in MATLAB)
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a Scatter plot using the MATLAB Method:
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
#Scatter plot using MATLAB method:
plt.scatter(y_axis, x_axis, c='red', label='Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title("Pyber Fare By Month")
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.legend()
```

```python
# Create a Bubble Charts Using MATLAB Method:
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# This code produces the same as the previous line. It's called List Comprehension and is basically like a 'for' loop.
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Creating a Scatter chart using Object-Oriented Interface:
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Making the previous Object-Oriented Interface scatter plot into a Bubble Chart:
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], linewidth=2, color='skyblue', edgecolors='black', label = 'Boston')
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
# Create a title.
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0,50)
ax.invert_yaxis()
ax.legend()
```

```python
# Create a Pie Chart Using the MATLAB Method:
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#getting crowded so increasing pie chart size, adding a wedge and details of perecentages
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface: (Lesson's SkillDrill)
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis, explode=explode_values, autopct='%1.1f%%',startangle=90,shadow=True, colors=colors)
fig.set_size_inches(8,8)
plt.show()
```
